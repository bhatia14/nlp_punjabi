counting words per sentence

```python
from nltk.corpus import inaugural
```

```python
inaugural.fileids()
```

```python
for speech in inaugural.fileids():
    words_total = len(inaugural.words(speech))
    print (words_total, speech)
```

```python
speech_len = [(len(inaugural.words(speech)), speech)for speech in inaugural.fileids()]
```

```python
speech_len
```

```python
max(speech_len)
```

```python
min(speech_len)
```

```python
for speech in inaugural.fileids():
    words_total = len(inaugural.words(speech))
    sents_total = len(inaugural.sents(speech))
    print(words_total/sents_total, speech)
```

Stemming/stemmers in nltk

```python
#porter Stemming
porter = nltk.PorterStemmer()
```

```python
my_list = ['cat', 'cats', 'lie', 'lying', 'run', 'running', 'city','cities', 'month', 'monthly', 'woman', 'women']
```

```python
for word in my_list:
    print(porter.stem(word))
```

```python
#lancaster Stem
lancaster = nltk.LancasterStemmer()
```

```python
for word in my_list:
    print(lancaster.stem(word))
```

```python
wnlem = nltk.WordNetLemmatizer()
```

```python
for word in my_list:
    print(wnlem.lemmatize(word))
```

Choices Exercise

```python
bs = nltk.corpus.gutenberg.words("bryant-stories.txt")
```

```python
bs_tags = nltk.pos_tag(bs, tagset="universal")
```

```python
bs_tags[:10]
```

```python
for((word1, tag1), (word2, tag2), (word3, tag3)) in nltk.trigrams(bs_tags):
    if tag1 == "NOUN" and word2 == "or" and tag3 == "NOUN":
        print(word1+" "+word2+" "+word3)
```

Chunking

```python
text = "I will go for coffee shop in New york after I get off from jet plane."
```

```python
text_tag = nltk.pos_tag(nltk.word_tokenize(text))
```

```python
text_tag
```

```python
sequence = '''
            Chunk:
            {<NNPS>+}
            {<NNP>+}
            {<NN>+}
            '''
```

```python
NPChunker = nltk.RegexpParser(sequence)
```

```python
result = NPChunker.parse(text_tag)
```

```python
print(result)
```
